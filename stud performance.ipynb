```python
data['Sex'].value_counts(normalize = True)
data['Sex'].value_counts(dropna = False).plot.bar(color = 'magenta')
plt.title('Comparison of Males and Females')
plt.xlabel('Sex')
plt.ylabel('count')
plt.show()
```

```python
data['Scholarship'].value_counts(normalize = True)
data['Reading'].value_counts(dropna = False).plot.bar(color = 'cyan')
plt.title('Comparison of various groups')
plt.xlabel('Groups')
plt.ylabel('count')
plt.show()
```

```python
data['Listening_in_Class'].value_counts(normalize = True)
data['Notes'].value_counts(dropna = False).plot.bar()
plt.title('Comparison of listening in class and notes')
plt.xlabel('Degree')
plt.ylabel('count')
plt.show()
```

```python
data['Grade'].value_counts(normalize = True)
data['Project_work'].value_counts(dropna = False).plot.bar(figsize = (18, 10))
plt.title('Comparison of grade and project work')
plt.xlabel('score')
plt.ylabel('count')
plt.show()
```
